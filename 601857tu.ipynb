```python
# -*- coding: utf-8 -*-

import matplotlib.pyplot as plt
from scipy.io import loadmat
import numpy as np
from sklearn.metrics import r2_score

def quota(real,pred,name):
    mape = np.mean(np.abs((pred - real) / real))
    # rmse
    rmse = np.sqrt(np.mean(np.square(pred - real)))
    # mae
    mae = np.mean(np.abs(pred - real))
    # R2
    r2 = r2_score(real, pred)
    print(name,'的mape:', mape, ' rmse:', rmse, ' mae:', mae, ' R2:', r2)


data01=loadmat('果实/601857LI-lstm_result.mat')['true']
data1=loadmat('果实/601857LI-lstm_result.mat')['pred']
data02=loadmat('果实/601857MS-lstm_result.mat')['true']
data2=loadmat('果实/601857MS-lstm_result.mat')['pred']
data03=loadmat('果实/601857lissa_lstm_result.mat')['true']
data3=loadmat('果实/601857lissa_lstm_result.mat')['pred']
data04=loadmat('果实/601857assa_lstm_result.mat')['true']
data4=loadmat('果实/601857assa_lstm_result.mat')['pred']

quota(data01,data1,'LSTM')
quota(data02,data2,'MS-LSTM')
quota(data03,data3,'SSA-LSTM')
quota(data04,data4,'MS-SSA-LSTM')
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.figure()

plt.subplot(2,2,1)
plt.rcParams.update({'font.size':7})
plt.plot(data01,c='r', label='Actual Price')
plt.plot(data1,c='b',label='Predicted Price')
plt.title('LSTM',fontsize=8)
plt.ylabel('Price')
#plt.xlabel('Day')
plt.xlabel('Day', horizontalalignment='right', x=1.0)
plt.legend()

plt.subplot(2,2,2)
plt.plot(data02,c='r', label='Actual Price')
plt.plot(data2,c='b',label='Predicted Price')
plt.title('MS-LSTM',fontsize=8)
plt.ylabel('Price')
#plt.xlabel('Day')
plt.xlabel('Day', horizontalalignment='right', x=1.0)
plt.legend()

plt.subplot(2,2,3)
plt.plot(data03,c='r', label='Actual Price')
plt.plot(data3,c='b',label='Predicted Price')
plt.legend()
plt.title('SSA-LSTM',fontsize=8)
plt.ylabel('Price')
#plt.xlabel('Day')
plt.xlabel('Day', horizontalalignment='right', x=1.0)
# In[7] 画图
plt.subplot(2,2,4)
plt.plot(data04,c='r', label='Actual Price')
plt.plot(data4,c='b',label='Predicted Price')
plt.legend()
plt.title('MS-SSA-LSTM',fontsize=8)
plt.ylabel('Price')
#plt.xlabel('Day')
plt.xlabel('Day', horizontalalignment='right', x=1.0)

plt.subplots_adjust(left=None,bottom=None,right=None,top=None,wspace=0.25,hspace=0.4)
plt.savefig('test4.jpg',dpi=600)
plt.show()
```
